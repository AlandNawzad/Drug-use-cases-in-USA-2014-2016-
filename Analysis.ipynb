```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots()
_ = plt.hist(df['AdmCount'], color = 'lightgreen')
```

```python
mean = df['AdmMonth'].mean()
mean
```

```python
xbar = df['AdmMonth'].mean()
```

```python
fig, ax = plt.subplots()
plt.axvline(xbar, color='orange')
_ = plt.hist(df['AdmMonth'], color = 'lightgray')
```

```python
median = np.median(df['AdmMonth']) 
```

```python
fig, ax = plt.subplots()
plt.axvline(mean, color='orange')
plt.axvline(median, color='green')
_ = plt.hist(mean, color = 'lightgray')
```

```python
varianse  = df['AdmMonth'].var()
varianse
```

```python
sigma = df['AdmMonth'].std()
sigma
```

```python
fig, ax = plt.subplots()
plt.axvline(varianse, color='orange')
plt.axvline(varianse+sigma, color='olivedrab')
plt.axvline(varianse-sigma, color='olivedrab')
_ = plt.hist(sigma, color = 'lightgray')
```
